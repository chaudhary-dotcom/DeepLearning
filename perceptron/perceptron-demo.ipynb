```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.sample(5)
```

```python
sns.scatterplot(data=df, x='petal_length', y='sepal_length', hue='species')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# x = df[['petal_length', 'sepal_length']]
df_filtered= df[df['species']!='setosa']

filter_df = df_filtered[['petal_length', 'sepal_length','species']]
filter_df.sample(5)
```

```python
x = filter_df[['petal_length', 'sepal_length']]
y=filter_df['species']
y.shape
```

```python
le = LabelEncoder()
y_encode = le.fit_transform(y)
y_encode.shape
x.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p.fit(x,y_encode)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
y_pred = p.predict(x)
```

```python
r2_score(y_encode,y_pred)
mean_absolute_error(y_encode,y_pred)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y_encode,clf=p,legend=2)
```
